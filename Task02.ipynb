```python
#identify continuous variable and discrete variable
def get_numerical_continuous(data, feas):
    numerical_continuous = []
    numerical_discrete = []
    for fea in feas:
        temp = data[fea].nunique()
        if temp <=10:
            numerical_discrete.append(fea)
        else:
            numerical_continuous.append(fea)
        
    return numerical_continuous, numerical_discrete
```

```python
numerical_continuous,numerical_discrete = get_numerical_continuous(data_train, numercial_fea)
```

```python
numerical_continuous
```

```python
numerical_discrete
```

```python
for item in numerical_discrete:
    print('{} has\n{}'.format(item, data_train[item].value_counts()))
```

```python
#check each column continuous variable data distribution
plt.figure(figsize=(9, 33*6))
for i in range(0,33):
    ax = plt.subplot(33,1,i+1)
    data_train[numerical_continuous[i]].plot(kind='hist', bins=100, density=True, ax=ax)
    ax.set_title(numerical_continuous[i])
```

As we observe from charts, loanAmnt is not normal distribution, and other features show skewness, specifically right skewed. We can use log funtion to transit feature and see if it fits normal distribution

```python
fig, ax = plt.subplots(1,2,figsize=(18,6), sharey=True)
ax[0].hist(data_train['loanAmnt'], bins=50)
ax[1].hist(np.log(data_train['loanAmnt']), bins=50)
```

```python
fig, ax = plt.subplots(4,2,figsize=(50,60))

for i in range(0,8):
    temp = numerical_discrete[i]
    ax = plt.subplot(4,2,i+1)
    sns.barplot(data_train[temp].value_counts(dropna=False).keys(), data_train[temp].value_counts(dropna=False),
            ax=ax)
    ax.set_title(numerical_discrete[i], fontsize=30)
    plt.yticks(fontsize=30)
    plt.xticks(fontsize=30)
```

As we can see, among discrete variables, applicationType, n11 and n12 have strong tendency. We will exclude those two from our next part.

```python
fig, ax = plt.subplots(3,2,figsize=(50,60))

for i in range(0,5):
    temp = categorical_fea[i]
    ax = plt.subplot(3,2,i+1)
    sns.barplot(data_train[temp].value_counts(dropna=False)[:20], data_train[temp].value_counts(dropna=False).keys()[:20], 
            ax=ax)
    ax.set_title(categorical_fea[i], fontsize=30)
    plt.yticks(fontsize=30)
    plt.xticks(fontsize=30)
```

```python
cond_0 = data_train['isDefault'] == 0
cond_1 = data_train['isDefault'] == 1
```

```python
plt.figure(figsize=(9, 41*6))
for i in range(0,41):
    ax = plt.subplot(41,1,i+1)
    data_train[numercial_fea[i]][cond_0].plot(kind='hist', density=True,bins=200, ax=ax,color='b')
    data_train[numercial_fea[i]][cond_1].plot(kind='hist', density=True,bins=200, ax=ax, color='r')
    ax.set_title(numercial_fea[i])
```

By comparing target value(0,1) on each numercial feature, some features show high overlap. Those may be removed when modeling.
n11, n12, n13, employeetitle show strong overlap areas.
